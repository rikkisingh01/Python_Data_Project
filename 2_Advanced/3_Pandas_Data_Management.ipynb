🟨 Fill Missing Salaries (2.3.3) - Problem
Pandas Data Management
Problem Statement:
Create a copy of the original DataFrame and fill in missing values in the salary_year_avg column with the median salary. Display the first 5 rows of the altered DataFrame.

```python
df_copy = df.copy()
#len(df_copy) == len(df)
median_salary_year_avg = df_copy['salary_year_avg'].median()
#you need to set the column to update to a new column, otherwise the dataframe is just a column
df_copy['salary_year_avg'] = df_copy['salary_year_avg'].fillna(median_salary_year_avg)
df_copy.head(5)
```
